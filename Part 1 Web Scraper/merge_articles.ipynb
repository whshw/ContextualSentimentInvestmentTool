```python
import matplotlib.pyplot as plt
import numpy as np

# 
countries = ['Australia', 'Canada', 'China', 'Denmark', 'Finland', 'France', 'Germany', 'Italy', 'Japan', 
             'Netherlands', 'New Zealand', 'Norway', 'Portugal', 'Singapore', 'South Korea', 'Spain', 
             'Sweden', 'Switzerland', 'United Kingdom', 'United States']

article_counts = [24722, 25573, 68070, 4766, 3961, 27785, 29165, 18603, 31090, 8415, 9967, 5729, 4399, 11105, 
                  16837, 14611, 6691, 8068, 17374, 60129]

# 
sorted_indices = np.argsort(article_counts)
sorted_countries = [countries[i] for i in sorted_indices]
sorted_counts = [article_counts[i] for i in sorted_indices]

# 
top5_countries = sorted_countries[-5:][::-1]
top5_counts = sorted_counts[-5:][::-1]
plt.figure(figsize=(12, 8))
plt.bar(top5_countries, top5_counts, color='#006284', width=0.5)
plt.xlabel('Country', fontsize=30)
plt.xticks(fontsize=28)
plt.ylabel('Article Count', fontsize=30)
plt.yticks(fontsize=28)
# plt.title('Article Count by Top 5 Countries', fontsize=25)
for i, count in enumerate(top5_counts):
    plt.text(i, count + 200, str(count), ha='center', fontsize=24)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.tight_layout()
plt.savefig('../Data/Images/article_count_by_top_countries.eps')
plt.show()

# 
bottom5_countries = sorted_countries[:5][::-1]
bottom5_counts = sorted_counts[:5][::-1]
plt.figure(figsize=(12, 8))
plt.bar(bottom5_countries, bottom5_counts, color='#006284', width=0.5)
plt.xlabel('Country', fontsize=30)
plt.xticks(fontsize=28)
plt.ylabel('Article Count', fontsize=30)
plt.yticks(fontsize=28)
# plt.title('Article Count by Bottom 5 Countries', fontsize=25)
for i, count in enumerate(bottom5_counts):
    plt.text(i, count + 50, str(count), ha='center', fontsize=24)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.tight_layout()
plt.savefig('../Data/Images/article_count_by_bottom_countries.eps')
plt.show()
```
